CLASSIFICATIONS MODEL

IMPORTING LIBRARIES

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score

```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=3)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
le=LinearRegression()
```

```python
le.fit(x_train,y_train)
```

```python
le.score(x_train,y_train)
```

```python
lepred=le.predict(x_test)
```

```python
print('predicted value and actual value',y_test,lepred)
```

```python
print('mean squared error',mean_squared_error(y_test,lepred))
print('mean absolute error',mean_absolute_error(y_test,lepred))
print('r2_score',r2_score(y_test,lepred))
```

Mean Squared Error (MSE) value is 9.062350069038218e-08 which indicates that, on average, the squared difference between the actual and predicted values is very small. Lower values of MSE indicate better performance.

Mean absolute error 0.00026409938729026193 is again very small, indicating that, on average, the absolute difference between the actual and predicted values is very small.

R2_score 0.9999999178333291 which is very close to 1, indicating that our model explains almost all of the variability of the dependent variable around its mean. In simpler terms, Our model fits the data extremely well
